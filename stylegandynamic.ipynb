```

LANDMARKS_MODEL_URL = 'http://dlib.net/files/shape_predictor_68_face_landmarks.dat.bz2'
hair_direction = np.load(f'/content/drive/My Drive/dlatent/hair_direction1.npy')
HoldLow = Image.open(f'/content/drive/My Drive/products/Hold_1.png')
HoldMed = Image.open(f'/content/drive/My Drive/products/Hold_3.png')
HoldHigh = Image.open(f'/content/drive/My Drive/products/Hold_4.png')
ShineYes1 = Image.open(f'/content/drive/My Drive/products/Yes_1.png')
ShineYes2 = Image.open(f'/content/drive/My Drive/products/Yes_2.png')
ShineNo1 = Image.open(f'/content/drive/My Drive/products/No_1.png')
ShineNo2 = Image.open(f'/content/drive/My Drive/products/No_2.png')
def unpack_bz2(src_path):
    data = bz2.BZ2File(src_path).read()
    dst_path = src_path[:-4]
    with open(dst_path, 'wb') as fp:
        fp.write(data)
    return dst_path


#if __name__ == "__main__":
    """
    Extracts and aligns all faces from images using DLib and a function from original FFHQ dataset preparation step
    python align_images.py /raw_images /aligned_images
    """
ts = 0
found = None
for file_name in glob.glob('/content/*.jpg'):
   fts = os.path.getmtime(file_name)
   if fts > ts:
      ts = fts
      found = file_name
if not os.path.isdir('ALIGNED_IMG'):
    os.mkdir('ALIGNED_IMG')
landmarks_model_path = unpack_bz2(get_file('shape_predictor_68_face_landmarks.dat.bz2',
                                           LANDMARKS_MODEL_URL, cache_subdir='temp'))
RAW_IMAGES_DIR = f'/content'
ALIGNED_IMAGES_DIR = f'/content/drive/My Drive/ALIGNED_IMG'

landmarks_detector = LandmarksDetector(landmarks_model_path)

for img_name in os.listdir(RAW_IMAGES_DIR):
    img_name=os.path.basename(found)
    #raw_img_path = os.path.join(RAW_IMAGES_DIR, img_name)   
    raw_img_path=found
    for i, face_landmarks in enumerate(landmarks_detector.get_landmarks(raw_img_path)):
        face_img_name = img_name
        aligned_face_path = os.path.join(ALIGNED_IMAGES_DIR, face_img_name)
        print('Saving image')
        image_align(raw_img_path, aligned_face_path, face_landmarks)
    break
imagename=''

def split_to_batches(l, n):
    for i in range(0, len(l), n):
        yield l[i:i + n]

src_dir = f'/content/drive/My Drive/ALIGNED_IMG'
dlatent_dir = f'/content/drive/My Drive/dlatent'
generated_images_dir = f'/content/drive/My Drive/generated_img'


ref_images = [os.path.join(src_dir, x) for x in os.listdir(src_dir)]
ref_images = list(filter(os.path.isfile, ref_images))
#ref_images=found1
if len(ref_images) == 0:
    raise Exception('%s is empty' % src_dir)

os.makedirs(generated_images_dir, exist_ok=True)
os.makedirs(dlatent_dir, exist_ok=True)

# Initialize generator and perceptual model

perceptual_model = PerceptualModel(image_size, layer=9, batch_size=batch_size)
perceptual_model.build_perceptual_model(generator.generated_image)

# Optimize (only) dlatents by minimizing perceptual loss between reference and generated images in feature space
for images_batch in tqdm(split_to_batches(ref_images, batch_size), total=len(ref_images)//batch_size):
    print(images_batch)
    names = [os.path.splitext(os.path.basename(x))[0] for x in images_batch]
    perceptual_model.set_reference_images(images_batch)
    op = perceptual_model.optimize(generator.dlatent_variable, iterations=iterations, learning_rate=lr)
    pbar = tqdm(op, leave=False, total=iterations)
    for loss in pbar:
        pbar.set_description(' '.join(names)+' Loss: %.2f' % loss)
    #print(' '.join(names), ' loss:', loss)

    # Generate images from found dlatents and save them
    generated_images = generator.generate_images()
    generated_dlatents = generator.get_dlatents()
    for img_array, dlatent, img_name in zip(generated_images, generated_dlatents, names):
        img = PIL.Image.fromarray(img_array, 'RGB')
        img.save(os.path.join(generated_images_dir, f'{img_name}.png'), 'PNG')
        imagename=os.path.join(dlatent_dir, f'{img_name}.npy')
        np.save(imagename, dlatent)

    generator.reset_dlatents()

def generate_image(latent_vector):
    latent_vector = latent_vector.reshape((1, 18, 512))
    generator.set_dlatents(latent_vector)
    img_array = generator.generate_images()[0]
    img = PIL.Image.fromarray(img_array, 'RGB')
    return img.resize((256, 256))
    
def move_and_show(latent_vector, direction, coeffs):
    images1=[]
    images=[]
   
    columns = 3
    
    for i, coeff in enumerate(coeffs):
        if coeff == 0:
          new_latent_vector = latent_vector.copy()
          new_latent_vector[:8] = (latent_vector + coeff*direction)[:8]
          img1 = generate_image(new_latent_vector)
          images1.append(img1)
        else:  
          new_latent_vector = latent_vector.copy()
          new_latent_vector[:8] = (latent_vector + coeff*direction)[:8]
          img = generate_image(new_latent_vector)
          images.append(img)
    grid = plt.GridSpec(4, 5, wspace=0.4, hspace=0.3)
    fig = plt.figure(figsize=(12,7)) 
    plt.subplot(grid[0:-1,0:2])
    plt.imshow(images1[0])
    plt.axis('off')
    
    x=0
    y=2
    z=3
    for i,image in enumerate(images):
      if(y>4):
        x=x+1
        y=2
        z=3
      plt.subplot(grid[x, y:z])
      y=y+1
      z=z+1 
      plt.imshow(image)
      plt.axis('off')
      
    plt.subplots_adjust(wspace=0, hspace=0, left=0, right=1, bottom=0, top=1)
    #plt.tight_layout()
    plt.show()
    
    def dropdownValueChanged():
      
      images_to_plot=[]
      
      if Hold.value!='Select' and Shine.value!='Select':
         
        if Hold.value == 'Low':
          images_to_plot.append(HoldLow)
          images_to_plot = list(filter(lambda x: x != HoldMed, images_to_plot))
          images_to_plot = list(filter(lambda x: x != HoldHigh, images_to_plot))

        elif Hold.value == 'Medium':
          images_to_plot.append(HoldMed)
          images_to_plot = list(filter(lambda x: x != HoldLow, images_to_plot))
          images_to_plot = list(filter(lambda x: x != HoldHigh, images_to_plot))

        elif Hold.value == 'High':
          images_to_plot.append(HoldHigh)
          images_to_plot = list(filter(lambda x: x != HoldMed, images_to_plot))
          images_to_plot = list(filter(lambda x: x != HoldLow, images_to_plot))
        elif Hold.value == 'Select':
          images_to_plot = list(filter(lambda x: x != HoldHigh, images_to_plot))
          images_to_plot = list(filter(lambda x: x != HoldMed, images_to_plot))
          images_to_plot = list(filter(lambda x: x != HoldLow, images_to_plot))

        if Shine.value == 'Yes':
          images_to_plot.append(ShineYes1)
          images_to_plot.append(ShineYes2)
          images_to_plot = list(filter(lambda x: x != ShineNo1, images_to_plot))
          images_to_plot = list(filter(lambda x: x != ShineNo2, images_to_plot))

        elif Shine.value == 'No':
          images_to_plot.append(ShineNo1)
          images_to_plot.append(ShineNo2)
          images_to_plot = list(filter(lambda x: x != ShineYes1, images_to_plot))
          images_to_plot = list(filter(lambda x: x != ShineYes2, images_to_plot))

        elif Shine.value == 'Select':
          images_to_plot = list(filter(lambda x: x != ShineNo1, images_to_plot))
          images_to_plot = list(filter(lambda x: x != ShineNo2, images_to_plot))
          images_to_plot = list(filter(lambda x: x != ShineYes1, images_to_plot))
          images_to_plot = list(filter(lambda x: x != ShineYes2, images_to_plot))

        plotImages(images_to_plot)

    
    def plotImages(images_to_plot):
      x=0
      y=2
      z=3
      
      fig = plt.figure(figsize=(30,20))
      grid1 = plt.GridSpec(4, 5)
      
      for image in images_to_plot:
        
        plt.subplot(grid1[x:-1,y:z])
        y=y+1
        z=z+1
        if(z>5):
          y=2
          z=3
        plt.imshow(image)
        plt.axis('off')
      
      
    def on_HoldSelection(b):       
      dropdownValueChanged()
      
    
    def on_ShineSelection(b):
      dropdownValueChanged()
      
        
    Hold = widgets.Dropdown(
    options=['Select','Low', 'Medium', 'High'],
    description='Hold:',
    disabled=False,
    )
    Shine = widgets.Dropdown(
    options=['Select','Yes', 'No'],
    description='Shine:',
    disabled=False,
    )
    
    
    #plt.subplot(grid1[0:-1,0:2])
    #plt.show(display(Hold))
    #plt.show(display(Shine))
    #plt.axis('off')
   
    
    display(Hold)
    display(Shine)
    '''HoldImage = []
    ShineImage = []
    HoldImage = Hold.observe(on_HoldSelection,names='value')
    ShineImage = Shine.observe(on_ShineSelection,names='value')'''
    Hold.observe(on_HoldSelection,names='value')
    Shine.observe(on_ShineSelection,names='value')


Stylegan = np.load(imagename)
move_and_show(Stylegan, hair_direction, [0.5,1,1.5,0,-0.5,-1,-1.5,-2,2,3] )
```
